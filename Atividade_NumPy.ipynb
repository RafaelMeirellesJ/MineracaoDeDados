Objetivo: O objetivo desta atividade é introduzir os conceitos básicos de mineração de dados e exploração de dados usando a biblioteca NumPy em Python. Os alunos irão aprender a manipular e analisar conjuntos de dados univariados e multivariados, calcular medidas de posição e dispersão, visualizar dados usando histogramas, scatter plots e box plots.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import scipy.stats as ss
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Mineração de Dados/iris.csv'
dados = pd.read_csv(file_path)
```

```python
dados.describe().T
```

```python
#cálculos de todos os valores de atributo da base de dado iris

#média
print(dados.median())
#percentil de 15%
print(dados.quantile(0.15))
#intervalo dos atributos
print(dados.max(numeric_only=True) - dados.min(numeric_only=True))
#variância
print(dados.var())
#desvio padrão  (em pandas, ddof=1 por padrão, ddof é o grau de liberdade)
print(dados.std())
#obliquidade
print(dados.skew())
#curtose
print(dados.kurtosis())
```

```python
#Fazendo histogramas
dados.hist(color = 'blue', grid=False, bins=20)
```

```python
#Criando dicionário mapeando as cores
classe_cor = {'Setosa' : 'red' , 'Virginica' : 'blue' , 'Versicolor' : 'green'}

#Criando uma lista com as cores
cores = [classe_cor[nome] for nome in dados.variety]

#Fazendo os scatter plots
scatter_matrix(dados, color=cores)
plt.show()
```

```python
#Fazendo scatter plot
dados.plot(kind = 'scatter', x = 'sepal.length', y = 'sepal.width', c = cores)
dados.plot(kind = 'scatter', x = 'petal.length', y = 'petal.width', c = cores)
plt.show()
```

```python
#Fazendo box plot
dados.plot(kind='box')
plt.show()
```
