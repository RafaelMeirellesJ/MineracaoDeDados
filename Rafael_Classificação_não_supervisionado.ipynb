# 1. Introdução
## O que é Classificação Não-Supervisionada?
A mineração de dados é uma área que visa descobrir informações valiosas a partir de conjuntos de dados complexos. A classificação é uma das tarefas essenciais da mineração de dados e é frequentemente dividida em duas categorias: **classificação supervisionada** e **classificação não supervisionada**.

**Classificação supervisionada** envolve a atribuição de rótulos (classes) a dados com base em exemplos previamente rotulados. Por outro lado, **classificação não supervisionada** lida com a segmentação de dados em grupos ou clusters com base em suas características, sem a necessidade de rótulos prévios.

A classificação não supervisionada é usada para explorar padrões ocultos nos dados e é amplamente aplicável em campos como análise de clientes, reconhecimento de padrões, agrupamento de documentos e muito mais.

# Diferença entre Classificação Não-Supervisionada e Supervisionada

A principal diferença entre classificação não supervisionada e supervisionada reside na presença ou ausência de rótulos nos dados:

*   Classificação Supervisionada: Os dados de treinamento incluem rótulos de classe conhecidos, permitindo que o modelo aprenda a mapear características para classes específicas.
*   Classificação Não-Supervisionada: Os dados não têm rótulos de classe; o objetivo é encontrar estrutura ou agrupamentos nos dados.

# Objetivos e Importância da Classificação Não-Supervisionada

A classificação não supervisionada desempenha um papel fundamental na mineração de dados e na análise de dados. Seus principais objetivos incluem:

*   **Descoberta de Padrões Ocultos**: Ajuda a identificar padrões, associações e estrutura intrínseca nos dados que podem não ser evidentes à primeira vista.
*   **Segmentação de Dados**: Permite a criação de grupos de dados com características semelhantes, o que é útil em tarefas como segmentação de clientes ou agrupamento de documentos.
*   **Redução de Dimensionalidade**: Pode ser usada para reduzir a complexidade de dados de alta dimensionalidade, tornando-os mais gerenciáveis.
*   **Visualização de Dados**: Facilita a visualização e interpretação de dados complexos.

A classificação não supervisionada é uma ferramenta poderosa na caixa de ferramentas de um cientista de dados e é amplamente aplicável em uma variedade de domínios.

# 2. Algoritmos de Classificação Não-Supervisionada

## Visão Geral dos Algoritmos

Existem vários algoritmos de classificação não supervisionada, cada um com suas próprias características e aplicações. Alguns dos algoritmos mais comuns incluem:

*   **K-means**: Divide os dados em *K clusters* com base na similaridade de seus atributos.
*   **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**: Encontra clusters de alta densidade de pontos e pode identificar pontos de dados como ruído.
*   **Agrupamento Hierárquico**: Cria uma hierarquia de clusters, que pode ser representada como uma árvore (dendrograma).
*   **Agrupamento Espectral**: Utiliza a matriz de afinidade dos dados para encontrar clusters.

Cada algoritmo tem suas próprias vantagens e limitações, e a escolha do algoritmo depende dos objetivos específicos da análise.

# O Algoritmo K-means

O algoritmo K-means é um dos algoritmos de classificação não supervisionada mais amplamente utilizados. Ele se destaca por sua simplicidade e eficiência. O objetivo do K-means é dividir um conjunto de dados em K clusters, onde cada cluster é representado por um centroide, que é o ponto médio dos membros do cluster.

# 3. Apresentação do Algoritmo K-means

## Conceito e Objetivo do K-means

O algoritmo K-means é amplamente utilizado em tarefas de classificação não supervisionada. Seu objetivo principal é dividir um conjunto de dados em grupos ou clusters com base em suas características, onde cada cluster é representado por um centroide.

O funcionamento básico do K-means pode ser resumido em três passos:

1.  Inicialização: Escolher aleatoriamente K centroides iniciais (pontos iniciais para os clusters).
2.   Atribuição de Clusters: Atribuir cada ponto de dados ao cluster cujo centroide é o mais próximo.
3.   Atualização dos Centroides: Recalcular os centroides com base nos membros atuais de cada cluster.

Esse processo de atribuição e atualização é repetido até que os centroides e as atribuições dos clusters não mudem significativamente ou um critério de parada seja atingido.

# Etapas Fundamentais do Algoritmo

O algoritmo K-means pode ser dividido em etapas fundamentais:

*   Inicialização dos Centroides: Os centroides iniciais podem ser escolhidos aleatoriamente ou com base em heurísticas. Uma inicialização adequada pode afetar significativamente os resultados.
*   Atribuição de Clusters: Cada ponto de dados é atribuído ao cluster cujo centroide está mais próximo. Isso é geralmente feito calculando a distância euclidiana entre os pontos e os centroides.
*   Atualização dos Centroides: Os centroides são recalculados com base nos pontos de dados atribuídos a cada cluster. Isso envolve calcular a média das características dos pontos no cluster.
*   Critério de Parada: O algoritmo continua iterando até que um critério de parada seja atingido. Isso pode ser um número máximo de iterações, convergência dos centroides ou outra condição.

## Como Determinar o Número de Clusters (K)?

Uma questão fundamental ao aplicar o K-means é determinar o número ideal de clusters (K). Escolher um K adequado é crucial, pois afeta a qualidade da segmentação.

Uma abordagem comum para determinar o K ideal é o método do cotovelo. Este método envolve calcular a inércia (ou a soma das distâncias quadráticas dos pontos aos centroides) para diferentes valores de K. O ponto em que a inércia começa a se estabilizar é o "cotovelo" e geralmente é escolhido como o número ideal de clusters.

# 4. Aplicação do K-means

## Casos de Uso Comuns

O algoritmo K-means é versátil e é amplamente aplicado em várias áreas, devido à sua eficiência e simplicidade. Alguns dos casos de uso comuns incluem:

*   Segmentação de Clientes: Empresas utilizam o K-means para agrupar clientes com base em seu comportamento de compra, facilitando a personalização de campanhas de marketing.
*   Análise de Imagens: Na área de visão computacional, o K-means é usado para segmentar imagens em regiões com características semelhantes.
*   Agrupamento de Documentos: Em processamento de linguagem natural, o K-means é empregado para agrupar documentos de texto semelhantes, auxiliando na categorização e recuperação de informações.
*   Biologia: Na biologia, o K-means é usado para classificar espécies com base em características genéticas.

# Importância da Métrica de Distância

Uma escolha crítica ao aplicar o K-means é a métrica de distância usada para calcular a similaridade entre pontos de dados e centroides. A métrica de distância mais comum é a distância euclidiana, que mede a distância direta entre dois pontos em um espaço euclidiano.

No entanto, dependendo do domínio do problema, outras métricas de distância, como a distância de Mahalanobis, podem ser mais apropriadas. A escolha da métrica de distância afeta diretamente os resultados do K-means, portanto, deve ser feita com cuidado.

# Avaliação da Qualidade dos Clusters

Uma vez que o K-means tenha sido aplicado e os clusters tenham sido formados, é importante avaliar a qualidade dos resultados. Algumas métricas comuns para avaliar a qualidade dos clusters incluem:

*   Inércia: A inércia mede a soma das distâncias quadráticas dos pontos de dados aos centroides dos clusters. Uma inércia menor indica que os pontos estão mais próximos de seus centroides, o que geralmente é desejável.
*   Silhueta: A pontuação de silhueta mede a coesão interna de um cluster em relação à separação entre os clusters. Ela varia de -1 a 1, com valores mais altos indicando uma separação melhor dos clusters.
*   Índice de Davies-Bouldin: Este índice considera a média das razões entre a dispersão intra-cluster e a dispersão inter-cluster. Valores mais baixos indicam clusters mais bem definidos.

# 5. Demonstração Prática

# Importação de Bibliotecas Relevantes

Antes de aplicar o algoritmo K-means em um conjunto de dados, é importante importar as bibliotecas relevantes. Você pode facilmente importar bibliotecas populares, como Pandas, Scikit-Learn e Matplotlib.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

# Seleção de Dados (Gerendo dados aleatórios)

Uma parte fundamental da análise de dados é a escolha dos dados a serem analisados. Como exemplo, usaremos o conjunto de dados de forma aleatória para a demonstração.

```python
# Gerar dados aleatórios
np.random.seed(0)
n_samples = 300
data = pd.DataFrame({
    'Compras Mensais': np.random.normal(100, 20, n_samples),
    'Número de Pedidos': np.random.randint(1, 6, n_samples),
    'Tempo Médio no Site (minutos)': np.random.normal(10, 2, n_samples)
})
```

# Pré-Processamento de Dados

Antes de aplicar o K-means, é importante realizar o pré-processamento dos dados. Isso inclui tarefas como lidar com valores ausentes, padronização de dados e seleção de recursos relevantes.

```python
# Padronizar os dados
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Visualizar as primeiras linhas do conjunto de dados
print(data_scaled[:5])
```

# Aplicação do Algoritmo K-means

Com os dados preparados, você pode aplicar o algoritmo K-means.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Determinar o número adequado de clusters (K) usando o método do cotovelo
inertia_values = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_scaled)
    inertia_values.append(kmeans.inertia_)

# Visualizar a curva de cotovelo
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia_values, marker='o', linestyle='--')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Inércia')
plt.title('Método do Cotovelo para Determinar o K Ideal')
plt.grid(True)
plt.show()
```

# Escolhendo o número de clusters (K)

```python
# Escolha o número de clusters (K)
k = 3

# Crie uma instância do modelo K-means
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)

# Treine o modelo com os dados padronizados
kmeans.fit(data_scaled)

# Obtenha as etiquetas dos clusters para cada cliente
labels = kmeans.labels_

# Calcular manualmente os centroides de cada cluster
import numpy as np

# Inicializar uma matriz para armazenar os centroides
centroids = np.zeros((k, data_scaled.shape[1]))

# Calcular os centroides manualmente
for cluster_label in range(k):
    cluster_data = data_scaled[labels == cluster_label]  # Selecionar dados pertencentes ao cluster
    cluster_centroid = np.mean(cluster_data, axis=0)  # Calcular a média das características do cluster
    centroids[cluster_label] = cluster_centroid

print("Coordenadas dos centroides:")
print(centroids)
```

# Calculando as distâncias entre os centroides

```python
from scipy.spatial import distance

# Calcular as distâncias entre os centroides
distances = np.zeros((k, k))

for i in range(k):
    for j in range(k):
        distances[i, j] = distance.euclidean(centroids[i], centroids[j])

print("Distâncias entre os centroides:")
print(distances)
```

# Visualização dos Resultados

A visualização é uma parte importante da análise de clusters. Você pode criar gráficos para representar os clusters gerados.

```python
plt.figure(figsize=(12, 8))

# Gráfico de dispersão com cores de cluster
plt.subplot(2, 1, 1)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, cmap='viridis', label='Clientes')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', label='Centroides', marker='X')
plt.xlabel('Compras Mensais (Padronizado)')
plt.ylabel('Número de Pedidos (Padronizado)')
plt.title('Segmentação de Clientes por Compras e Número de Pedidos com Centroides')
plt.legend()

# Matriz de distâncias
plt.subplot(2, 1, 2)
plt.imshow(distances, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Matriz de Distâncias entre Centroides')
plt.xlabel('Clusters')
plt.ylabel('Clusters')

plt.tight_layout()
plt.show()
```

Algoritmo não supervisionado é o algoritmo que não possui uma classe alvo, ou seja, não sabemos o que o modelo deseja prever.

Geralmente os algoritmos não supervisionados geram clusters (grupos de registros semelhantes) que após gerados, ainda é necessário analisar o que o cluster representa.

```python
from sklearn.cluster import KMeans
#from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Realiza a leitura dos dados
df = pd.read_csv('iris.csv')

# Separa apenas os atributos para o treinamento (atributos são as "colunas" previsoras)
X = df[['comprimento sepala', 'largura sepala', 'comprimento petala', 'largura petala']].values
```

## Como calcular a quantidade de clusters?
Neste caso já sabemos que a quantidade de clusters é 3, mas caso não soubéssemos, usaríamos o método do cotovelo, conforme vems abaixo

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from IPython.display import clear_output

# Defina o intervalo de valores k
k_range = range(1, 10)

# Execute o algoritmo de clustering para cada valor k e armazene a soma dos erros quadráticos em uma lista
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(X)
    sse.append(km.inertia_)
    # Limpa as mensagens de saída
    clear_output(wait=True)

# Trace a curva de cotovelo
plt.plot(k_range, sse)
plt.xlabel('Número de Clusters (k)')
plt.ylabel('soma dos erros quadráticos ')
plt.title('Método do Cotovelo')
plt.show()
```

## Treinar / exibir clusters

```python
# Treinando o modelo de K-means com 3 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
# Visualizando os clusters criados
plt.scatter(X[:, 2], X[:, 3], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], marker='x', s=200, linewidths=3, color='r')
plt.xlabel("Comprimento Pétala")
plt.ylabel("Largura Pétala")
plt.title('Largura x Comprimento Pétala')
plt.show()

# Visualizando os clusters criados
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, linewidths=3, color='r')
plt.xlabel("Comprimento Sépala")
plt.ylabel("Largura Sépala")
plt.title('Largura x Comprimento Sépala')
plt.show()
```
